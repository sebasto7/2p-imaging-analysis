# Copying pData files of a list from one dolfer to another

This script was generated to help me sort pData files for the Molina-Obando et al., 2019 paper into different figures.

```python
import os
import shutil
import pandas as pd

# Path to the 2p_pData folder where the files are located
source_folder = r"D:\NAS Server seb backup\2p data Investigator\RedHP\2p_pData"

# Path to the output folder where the selected files will be copied
output_folder = r"D:\NAS Server seb backup\2p data Investigator\RedHP\out_2pData"

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# Read the Excel file (replace 'your_file.xlsx' with the path to your file)
excel_path = r"D:\NAS Server seb backup\2p data Investigator\RedHP/temp_pData_list.xlsx"
df = pd.read_excel(excel_path)

# Loop through the 'fname' column to get the file names
for fname in df['fname']:
    # Full path of the source file
    src_file = os.path.join(source_folder, fname)
    
    # Check if the file exists in the source directory
    if os.path.exists(src_file):
        # Full path of the destination file
        dst_file = os.path.join(output_folder, fname)
        
        # Copy the file to the output directory
        shutil.copy(src_file, dst_file)
        print(f"Copied: {fname}")
    else:
        print(f"File not found: {fname}")
```
